# 라이브러리 추가

```python
# 라이브러리 추가

import pandas as pd
```

```python
# 2025년 데이터 가져오기

testCSV = "kamis_prices_2025_verification_no_duplicates.csv"
test_df = pd.read_csv(testCSV)
```

```python
test_df.head()
```

```python
test_df.shape
```

```python
columns_for_grouping = ['MY_TARGET_ITEM_NAME','date']

df_daily_avg_price = test_df.groupby(
        by=['MY_TARGET_ITEM_NAME','date'],
        sort=False
    )['price_api'].mean().reset_index()

df_daily_avg_price.head()
```

```python
df_daily_avg_price.tail()
```

```python
grouping_keys = ['date', 'MY_TARGET_ITEM_NAME']

df_daily_avg_price = df_daily_avg_price.groupby(grouping_keys)['price_api'].mean().reset_index()
```

```python
df_daily_avg_price
```

```python
pivot_column_name = 'MY_TARGET_ITEM_NAME'

try:
    df_daily_avg_price = df_daily_avg_price.pivot_table(index = ['date'],
                                    columns = pivot_column_name,
                                    values = 'price_api',
                                    aggfunc = 'mean')
except KeyError as e:
    print(f"Check Column's name")
except Exception as e:
    print(f"Other error happened")
```

```python
df_daily_avg_price.columns.name = None
df_daily_avg_price = df_daily_avg_price.reset_index()
df_daily_avg_price = df_daily_avg_price.round(2)
```

```python
df_daily_avg_price
```

```python
column_order = [
        'date',
        '배추',
        '무',
        '마늘',
        '양파',
        '대파',
        '홍고추',
        '깻잎'
    ]
df = df_daily_avg_price[column_order]
```

```python
df
```

```python
item_en = {
    'date':'date',
    '배추': 'cabbage',
    '무': 'radish',
    '양파': 'onion',
    '마늘': 'garlic',
    '대파': 'daikon',
    '홍고추': 'cilantror',
    '깻잎': 'artichoke'
}

df.columns = df.columns.map(item_en)
```
